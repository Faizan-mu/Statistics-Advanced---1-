 **There are two main types of random variables:**

 **Discrete Variables**- A discrete random variable can take on a finite or countably infinite set of distinct values, such as the number of goals in a football match or the number of students present in class.

**Question 2) What are the types of random variables?**

**Answer 2)** Random variables are mainly classified into two types based on the type of values they can take-

**1. Discrete Random Variable**

- A discrete random variable takes countable values.

- These values can be finite or countably infinite (like 0, 1, 2, 3, ...).

**Examples:**

- Number of heads when tossing 3 coins.

- Number of students present in class.

- Number of cars passing a toll booth in one hour.

**Key Features:**

- Values are separated and distinct.

- Probability can be assigned to each specific value using a probability mass function (PMF).

**2. Continuous Random Variable**-

A continuous random variable takes uncountably infinite values within an interval or range.

These values are often measurable quantities.

**Examples:**

- Height or weight of a person.

- Time taken to finish a race.

- Temperature on a given day.

**Key Features:**

- Values form a continuous scale.

- Probability is measured over an interval using a probability density function (PDF).

- Probability of taking an exact value is zero, but probability over a range is positive (e.g., P(5 < X < 10)).


**Question 3) Explain the difference between discrete and continuous distributions.**

**Answer 3)** In probability theory, discrete and continuous distributions describe how the values of random variables are distributed, but they differ based on the type of values the variable can take.

**Discrete distribution** It can applies to a discrete random variable, which takes on specific, countable values. The probabilities of these values are represented using a probability mass function (PMF). Each outcome has a distinct, non-zero probability, and the total sum of all probabilities is equal to 1. Examples of discrete distributions include the Binomial distribution, which models the number of successes in a fixed number of trials, and the Poisson distribution, which models the number of events occurring in a fixed interval of time or space.

**Continuous distribution** It can applies to a continuous random variable, which can take on any value within a range or interval. Instead of individual probabilities, it uses a probability density function (PDF) to describe the likelihood of outcomes. Since there are infinitely many possible values, the probability of the variable taking any exact value is zero; instead, we calculate the probability over an interval (e.g., P(2 < X < 5)). Common continuous distributions include the Normal distribution (bell-shaped curve used in statistics) and the Exponential distribution (used to model time until an event occurs).

**Question 7) What is the significance of confidence intervals in statistical analysis?**

**Answer 7) A confidence interval (CI)** is a range of values, derived from sample data, that is used to estimate an unknown population parameter, such as the mean or proportion. It provides not just a single estimate (like a sample mean), but a range within which we are reasonably certain the true value lies. The confidence level (commonly 90%, 95%, or 99%) indicates how confident we are that this range contains the true parameter.



**Significance of Confidence Intervals:**

**1) Provides a Range, Not Just a Point**

Instead of guessing one value (e.g., population mean = 70), a CI gives a range (e.g., 67 to 73), which is a more informative and realistic estimate.

**2) Measures Uncertainty**

Confidence intervals reflect the degree of uncertainty or variability in the estimate. A wider interval means more uncertainty; a narrower interval indicates more precision.

**3) Supports Decision-Making**

In research, business, and science, CIs help assess the reliability of sample results. For example, if a 95% CI for a drug's effect does not include zero, it may indicate the drug is effective.

**4) Used in Hypothesis Testing**

Confidence intervals can be used as an alternative to p-values in hypothesis testing. If the hypothesized value (e.g., mean = 50) falls outside the CI, we may reject the null hypothesis.

**5) Applies Across Fields**

Whether in medicine (estimating effectiveness of treatment), economics (estimating unemployment rates), or manufacturing (estimating defect rates), CIs provide a standard method for expressing statistical conclusions.

**Question 8) What is the concept of expected value in a probability distribution?**

**Answer 8)**The expected value (also called the mean or mathematical expectation) of a probability distribution is the average value we would expect to obtain if an experiment were repeated many times. It represents the long-run average outcome of a random variable.

Mathematically, the expected value is a weighted average of all possible values that a random variable can take, where each value is weighted by its probability.

```python
""" Question 9) Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution. """

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print(f"Calculated Mean: {mean:.2f}")
print(f"Calculated Standard Deviation: {std_dev:.2f}")

# Step 3: Plot the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean = {mean:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
""" Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
‚óè Write the Python code to compute the mean sales and its confidence interval. """

import numpy as np
import scipy.stats as stats

# Sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]


mean_sales = np.mean(daily_sales)
std_dev_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

standard_error = std_dev_sales / np.sqrt(n)


z_value = 1.96


margin_of_error = z_value * standard_error


lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Output results
print(f"Sample Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
